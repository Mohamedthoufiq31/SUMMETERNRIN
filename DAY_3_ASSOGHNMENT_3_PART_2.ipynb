Q12.   Create a random dataset of 100 rows and 30 columns. All the values are defined between [1,200].  Perform
 	the following operations:
(i)	Replace all the values with NA in the dataset defined between [10, 60]. Print the count of number rows having missing values.
(ii)	Replace all the NA values with the average of the column value.  
(iii)	Find the Pearson correlation among all the columns and plot heat map. Also select those columns having correlation <=0.7.
(iv)	Normalize all the values in the dataset between 0 and 10.
(v)	Replace all the values in the dataset with 1 if value <=0.5 else with 0.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)  # For reproducibility
data = np.random.randint(1, 201, size=(100, 30))
df = pd.DataFrame(data, columns=[f'col_{i+1}' for i in range(30)])
```

```python
df.replace(list(range(10, 61)), np.nan, inplace=True)
rows_with_na = df.isna().any(axis=1).sum()
print(f'Number of rows with missing values: {rows_with_na}')
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
# Select columns with correlation <= 0.7
selected_columns = correlation_matrix.columns[(correlation_matrix <= 0.7).any()]
print(f'Columns with correlation <= 0.7: {selected_columns.tolist()}')
```

```python
df_normalized = (df - df.min()) / (df.max() - df.min()) * 10
```

```python
df_binary = df_normalized.applymap(lambda x: 1 if x <= 0.5 else 0)
```

```python
# Display final dataset
print(df_binary.head())
```

Q13.
 	Create a random dataset of 500 rows and 10 columns.  
Columns 1 to 4 are defined between [-10, 10];  Columns 5 to 8 are defined between [10, 20];   Columns 9 to 10 are defined between [-100, 100].  
Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric graph using each algorithm. (i) 	K-Mean clustering
(ii) Hierarchical clustering


```python
import pandas as pd
import numpy as np
np.random.seed(42)
data = np.hstack([
    np.random.uniform(-10, 10, size=(500, 4)),
    np.random.uniform(10, 20, size=(500, 4)),
    np.random.uniform(-100, 100, size=(500, 2))
])
df = pd.DataFrame(data, columns=[f'Col{i+1}' for i in range(10)])
print(df.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
Z = linkage(df, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(Z, truncate_mode='lastp', p=30)
plt.xlabel('Number of points in node (or index of point if no parenthesis)')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
last = Z[-10:, 2]
last_rev = last[::-1]
indexes = np.arange(1, len(last) + 1)

plt.figure(figsize=(8, 5))
plt.plot(indexes, last_rev, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Distance')
plt.title('Distance Metric For Optimal Number of Clusters (Hierarchical)')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate the dataset
np.random.seed(42)  # For reproducibility
data = np.hstack([
    np.random.uniform(-10, 10, size=(500, 4)),   # Columns 1 to 4
    np.random.uniform(10, 20, size=(500, 4)),    # Columns 5 to 8
    np.random.uniform(-100, 100, size=(500, 2))  # Columns 9 to 10
])

# Create a DataFrame
df = pd.DataFrame(data, columns=[f'Col{i+1}' for i in range(10)])
print(df.head())

# K-Means Clustering
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)

# Plot the elbow graph for K-Means
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method For Optimal k')
plt.show()

# Hierarchical Clustering
Z = linkage(df, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z, truncate_mode='lastp', p=30)
plt.xlabel('Number of points in node (or index of point if no parenthesis)')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()

# Plot distance metric graph (elbow method for hierarchical clustering)
last = Z[-10:, 2]
last_rev = last[::-1]
indexes = np.arange(1, len(last) + 1)

plt.figure(figsize=(8, 5))
plt.plot(indexes, last_rev, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Distance')
plt.title('Distance Metric For Optimal Number of Clusters (Hierarchical)')
plt.show()
```

Q15. Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].  
Perform the following operations:
(i)	Perform t-Test on each column.
(ii)	Perform Wilcoxon Signed Rank Test on each column.
(iii)	Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4


```python
import pandas as pd
import numpy as np
np.random.seed(42)
data = np.random.uniform(5, 10, size=(500, 5))
df = pd.DataFrame(data, columns=[f'Col{i+1}' for i in range(5)])
print(df.head())
```

```python
from scipy.stats import ttest_1samp
t_test_results = {}
for col in df.columns:
    t_stat, p_value = ttest_1samp(df[col], 7.5)
    t_test_results[col] = (t_stat, p_value)

print("t-Test results (t-statistic, p-value):")
for col, (t_stat, p_value) in t_test_results.items():
    print(f"{col}: t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
```

```python
from scipy.stats import wilcoxon
wilcoxon_results = {}
for col in df.columns:
    w_stat, p_value = wilcoxon(df[col] - 7.5)
    wilcoxon_results[col] = (w_stat, p_value)
print("Wilcoxon Signed Rank Test results (W-statistic, p-value):")
for col, (w_stat, p_value) in wilcoxon_results.items():
    print(f"{col}: W-statistic = {w_stat:.3f}, p-value = {p_value:.3f}")
```

```python
from scipy.stats import ttest_ind, ranksums
t_stat, p_value = ttest_ind(df['Col3'], df['Col4'])
print(f"Two Sample t-Test between Column 3 and Column 4: t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
w_stat, p_value = ranksums(df['Col3'], df['Col4'])
print(f"Wilcoxon Rank Sum Test between Column 3 and Column 4: W-statistic = {w_stat:.3f}, p-value = {p_value:.3f}")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums
np.random.seed(42)
data = np.random.uniform(5, 10, size=(500, 5))
df = pd.DataFrame(data, columns=[f'Col{i+1}' for i in range(5)])
print(df.head())
t_test_results = {}
for col in df.columns:
    t_stat, p_value = ttest_1samp(df[col], 7.5)
    t_test_results[col] = (t_stat, p_value)
print("\nt-Test results (t-statistic, p-value):")
for col, (t_stat, p_value) in t_test_results.items():
    print(f"{col}: t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
wilcoxon_results = {}
for col in df.columns:
    w_stat, p_value = wilcoxon(df[col] - 7.5)
    wilcoxon_results[col] = (w_stat, p_value)
print("\nWilcoxon Signed Rank Test results (W-statistic, p-value):")
for col, (w_stat, p_value) in wilcoxon_results.items():
    print(f"{col}: W-statistic = {w_stat:.3f}, p-value = {p_value:.3f}")
t_stat, p_value = ttest_ind(df['Col3'], df['Col4'])
print(f"\nTwo Sample t-Test between Column 3 and Column 4: t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
w_stat, p_value = ranksums(df['Col3'], df['Col4'])
print(f"Wilcoxon Rank Sum Test between Column 3 and Column 4: W-statistic = {w_stat:.3f}, p-value = {p_value:.3f}")
```
