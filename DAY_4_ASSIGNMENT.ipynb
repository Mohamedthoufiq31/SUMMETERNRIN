```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Wine dataset.csv')
df.head()
```

```python
duplicate = df.duplicated()
print(duplicate.sum())
df[duplicate]
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df.shape
```

```python
df['class'].value_counts().plot(kind = 'pie' , autopct='%0.2f')
```

```python
df['Malic acid'] = [1 if val >=7 else 0 for val in df['Malic acid']]
```

```python
df['quality'].value_counts().plot(kind = 'bar')
plt.show()
```

```python
df['Malic acid'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 10))

# Exclude 'Malic acid' from the features to plot
features = ['class', 'Alcohol', 'Ash', 'Proline ']
n_features = len(features)

# Creating subplots for each feature
for i in range(n_features):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df, x=features[i], hue='class', kde=True)
    plt.title(features[i])

plt.tight_layout()
plt.show()
```
